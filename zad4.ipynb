```python
%load_ext autoreload
%autoreload 2
%matplotlib widget

import matplotlib.pyplot as plt
from dgeometry import *
import dgeometry as dg
Entity.displayMethod = 'Cords'
class zad4:
    A = Point(3,4.5,1.5)('A')
    B = Point(1.5,14,3)('B')
    O = Point(8,10,9)('O')
    D = Point(10,7.5,2)('D')
    p1 = Point(3,13,4.5)('1')
    p2 = Point(0,15,1.5)('2')
    p3 = Point(5,6.65,4.4)('3')
    p4 = Point(2.5,7.65,2)('4')
    P = Point(6.2,6.15,5.4)('P')
    H = Point(1.5,4.5,9.5)(' ')
    ao = (A^O)(' ')
    bo = (B^O)(' ')
    ab = (A^B)(' ')
    def __init__(self):
        self.A.plot_all()
        (self.A@VPP).plot_vp()
        (self.A@HPP).plot_hp()
        self.B.plot_all()
        (self.B@VPP).plot_vp()
        (self.B@HPP).plot_hp()
        self.O.plot_all()
        (self.O@VPP).plot_vp()
        (self.O@HPP).plot_hp()
        self.ab.plot_all()
        (self.ab@VPP).plot_vp()
        (self.ab@HPP).plot_hp()
        self.bo.plot_all()
        (self.bo@VPP).plot_vp()
        (self.bo@HPP).plot_hp()
        self.ao.plot_all()
        (self.ao@VPP).plot_vp()
        (self.ao@HPP).plot_hp()
        (self.D@VPP).plot_vp()
        (self.D@HPP).plot_hp()
        self.D.plot_all()
    def showP(self):
        (self.P@VPP).plot_vp()
        (self.P@HPP).plot_hp()
        (self.P).plot_all()
    def conA1bis(self):
        ((self.A^self.p1)('horizontal')@VPP).plot_vp()
        #((self.A^self.p1)('horizontal')).plot_all()
        (self.p1@VPP).plot_vp()
    def horizontalPlane(self):
        ((self.A^self.p1)('horizontal')@HPP).plot_hp()
        #((self.A^self.p1)('horizontal')).plot_all()
        (self.p1@HPP).plot_hp()
    def verticalPlane(self):
        ((self.A^self.p2)('vertical')@VPP).plot_vp()
        #((self.A^self.p2)('vertical')).plot_all()
        (self.p2@VPP).plot_vp()
    def conA2prim(self):
        ((self.A^self.p2)('vertical')@HPP).plot_hp()
        #((self.A^self.p2)('vertical')).plot_all()
        (self.p2@HPP).plot_hp()
    def showPerpendiculars(self):
        ((self.H^self.D)('perpendicular')@HPP).plot_hp()
        ((self.H^self.D)('perpendicular')@VPP).plot_vp()
        ((self.H^self.D)('perpendicular')).plot_all()
    def show34(self):
        (self.p3@VPP).plot_vp()
        (self.p3@HPP).plot_hp()
        (self.p4@VPP).plot_vp()
        (self.p4@HPP).plot_hp()
    def con34(self):
        ((self.p4^self.P)(' ')@HPP).plot_hp()
        ((self.p4^self.P)(' ')).plot_all()

SETUP3D = (20,60) # something like perspective ;)

GeometrySceneDG(init_3d=SETUP3D,figsize=(16,15) ) 
zad = zad4()
zad.horizontalPlane()
zad.conA1bis()
zad.verticalPlane()
zad.conA2prim()
zad.showPerpendiculars()
zad.show34()
zad.con34()
zad.showP()



```
